### Problem 5: The function np.mod(a,b) in the numpy library is the modulo of a with respect to b. Ths is the remainder of a when divided by b.

#### a. Using a loop print all the numbers between 1 and 100 which are evenly divisible by 5.

```python
list5 = []
for i in range(1,101):
    if np.mod(i,5) == 0:
        list5.extend([i])
print("All the numbers between 1 and 100 which are evenly divisible by 5 are " + str(list5) + ".")
```

#### b. Now repeat the same procedure, but without a for loop.
